## **Hotel Booking Cancellation Prediction**

**1. Descripción del problema**

Las cancelaciones de reservas representan un problema importante para la industria hotelera. Cuando un cliente cancela una reserva, el hotel pierde ingresos potenciales y tiene dificultades para optimizar la ocupación.

El objetivo de este proyecto es predecir si una reserva será cancelada (is_canceled) antes de la fecha de llegada, utilizando técnicas de Machine Learning.

Este tipo de modelo puede ayudar a los hoteles a:

- Optimizar estrategias de overbooking

- Ajustar precios dinámicamente

- Detectar reservas con alto riesgo de cancelación

- Mejorar la planificación de ocupación


**2. Dataset utilizado**

Se ha utilizado el dataset Hotel Booking Demand Dataset en Kaggle que contiene que contiene información sobre reservas de hoteles.

    [Visitar Kaggle](https://www.kaggle.com/datasets/jessemostipak/hotel-booking-demand)


Características principales del dataset:

- Contiene información sobre reservas en hoteles urbanos y resorts

Incluye variables como:

- país de origen

- tiempo de antelación de la reserva (lead_time)

- número de adultos

- cancelaciones anteriores

- peticiones especiales

- agente o canal de reserva

El dataset incluye variables:

- Numéricas (por ejemplo: lead_time, previous_cancellations)

- Categóricas (por ejemplo: country, hotel, market_segment)

- Variable objetivo:

is_canceled

1 : la reserva fue cancelada

0 : la reserva no fue cancelada



**3. Solución adoptada**

Se ha desarrollado un pipeline de Machine Learning para clasificación binaria.

El flujo de trabajo seguido ha sido:

- Limpieza del dataset dee columnas con muchos nulos y limpieza de filas con nulos.

- Análisis exploratorio de datos (EDA)

        -Análisis de variables numéricas y categóricas

        -Relación con la variable objetivo is_canceled

  

- Agrupación de valores de columnas con dispersion de valores en pocas clases

- Reducción de cardinalidad en variables categóricas

- Preprocesamiento de categoricas

- Codificación de variables categóricas

Pipeline de preprocesado con ColumnTransformer

Entrenamiento de modelos

        -Random Forest

        -Regresion Logistica

        -Lightgbm

        -XGBoost

        -CatBoost
        

Evaluación del modelo con confusion matrix para elegir el mejor modelo

Mejora de hiperparametros con GrisSearch

Obtención de Accuracy
