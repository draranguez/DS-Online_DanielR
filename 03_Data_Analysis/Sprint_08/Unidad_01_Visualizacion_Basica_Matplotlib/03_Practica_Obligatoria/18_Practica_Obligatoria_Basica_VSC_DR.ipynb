Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:

Los colores de las barras deben tener un 50% de transparencia y ser: azul, gris, rojo, aceituna y marrn.  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Arial", tamao = 14, peso = "Bold", color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.
4. El boxplot de la altura de las ciudades.

###SOLUCION FIGURA 1 POR SEPARADO

```python
def separa_nombre(nombre):
    nuevo=nombre
    for letra in nombre[1:]:
        if letra == letra.upper():
            nuevo +=" "
            nuevo += letra
    return nuevo

df_ciudades_mas_pobladas= df[["city","population_total"]].nlargest(5, columns="population_total").copy()
df_ciudades_mas_pobladas= df_ciudades_mas_pobladas.set_index("city")

ciudades = [separa_nombre(ciudad) for ciudad in df_ciudades_mas_pobladas.index] #list comprehension to separate names,sabiendo que el index ya se ha configurado como el nombre de la propia ciudad

colores= ["blue","gray","red","olive","brown"]

plt.figure(figsize=(8,6))

container =plt.bar(x=ciudades,height= df_ciudades_mas_pobladas["population_total"]/1000, color=colores, alpha=0.7)


plt.xlabel("Ciudades")
plt.ylabel("Poblacin total (en miles)")
plt.ylim([0,5000])
plt.title = ("Top 5 ciudades ms pobladas de California")

plt.bar_label(container, labels=[f"{int(valor):,}" for valor in df_ciudades_mas_pobladas["population_total"]/1000], padding=3)

plt.show()

```

###Solucion Figuura 2 por separado

Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Arial", tamao = 14, peso = "Bold", color azul acero o "steelblue")

```python
plt.figure(figsize=(8,7))


plt.hist(df["area_total_km2"],
         color= "steelblue",
         alpha= 0.7,
         bins= 50,  #los bins ya agrupan por "nmero de ciudades" dentro de los 50 rangos
         label= "Hist Area")

fontdict = dict(family="Arial", size= 14, weight="bold") #hace referencia a las fuentes para etiquetas y titulos

plt.xlabel("Superficie ($km^2$)", fontdict= fontdict)
plt.ylabel("Nmero de Ciudades", fontdict= fontdict)
plt.title("Histograma del rea total ocupada por las ciudades", fontdict= fontdict)
plt.legend()

plt.show()
```

###SOL FIGURA 3 POR SEPARADO
El scatter de rea de agua de un ciudad con su poblacion. Transparencia al 50% y color verde.

```python
plt.figure(figsize=(8,6))  #tamao del grfico

#Tipo grafico
plt.scatter(x=df["area_water_km2"],
            y=df["population_total"]/1000,
            alpha=0.5,
            color="green")

plt.xlabel("rea de agua ($km^2$)")
plt.ylabel("Poblacin total (en miles)")

#Ahora vayamos con las anotaciones outliers

outlier_1_y=df_ciudades_mas_pobladas["population_total"].nlargest(2).iloc[0]
outlier_2_y=df_ciudades_mas_pobladas["population_total"].nlargest(2).iloc[1]
outlier_1_x=df.loc[df["population_total"]==outlier_1_y, "area_water_km2"].values[0]
outlier_2_x=df.loc[df["population_total"]==outlier_2_y, "area_water_km2"].values[0]

text="Outliers:\nLos ngeles y San Diego"
xytext=(60,3000)  #position of the text box


plt.annotate( xy=(outlier_1_x, outlier_1_y/1000),
            xytext=xytext,
            text=text,
            arrowprops=dict(arrowstyle ="->", color="red"),
            fontsize=12,
            color="red")

plt.annotate(xy=(outlier_2_x, outlier_2_y/1000),
            xytext=xytext,
            text=" "*len(text),  #to avoid overlapping of texts
            arrowprops=dict(arrowstyle ="->", color="red"),
            fontsize=12,
            color="red")

plt.title("grafica")

plt.show()

```

FIGURA 4 POR SEPARADO
BOXPLOT CON LA ALTURA DE LAS CIUDADES

```python
plt.figure(figsize=(8,6))

plt.boxplot(df["elevation_m"],
            vert= False,
            patch_artist= True, # Sin patch_artist=True no nos aceptar las ediciones posteriores0
            boxprops = dict(facecolor = "red"),
            medianprops = dict(color= "black"))

plt.xlabel("Altitud (m)")
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.yticks([]) # Con yticks podemos quitar el ticker del eje y (el 1 que sala por defecto)

plt.show()
```
